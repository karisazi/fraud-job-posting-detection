```python
from transformers import MarianMTModel, MarianTokenizer

def translate(text, src_lang, tgt_lang):
    model_name = f"Helsinki-NLP/opus-mt-{src_lang}-{tgt_lang}"
    tokenizer = MarianTokenizer.from_pretrained(model_name)
    model = MarianMTModel.from_pretrained(model_name)

    tokenized = tokenizer(text, return_tensors="pt", padding=True)
    translated = model.generate(**tokenized)
    return tokenizer.decode(translated[0], skip_special_tokens=True)

# Example back-translation: English → French → English
text = "I love this movie."
fr_text = translate(text, "en", "it")
en_text = translate(fr_text, "it", "en")
print(en_text)
```

```python
import random
import pandas as pd
from typing import List
import torch
from tqdm import tqdm
from transformers import MarianMTModel, MarianTokenizer
from sentence_transformers import SentenceTransformer, util

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# Map pivot aliases to MarianMT codes
LANG_CODE_MAP = {
    'jap': 'ja',   # Japanese
    'ja': 'ja',
    'zh': 'zh',    # Chinese simplified
    'el': 'el',    # Greek
    # others (de, fr, es, id, etc.) stay as-is
}

CACHE = {}
def get_translator(src, tgt):
    src_code = LANG_CODE_MAP.get(src, src)
    tgt_code = LANG_CODE_MAP.get(tgt, tgt)
    key = (src_code, tgt_code)
    if key not in CACHE:
        name = f"Helsinki-NLP/opus-mt-{src_code}-{tgt_code}"
        tok = MarianTokenizer.from_pretrained(name)
        mdl = MarianMTModel.from_pretrained(name).to(DEVICE)
        mdl.eval()
        CACHE[key] = (tok, mdl)
    return CACHE[key]

@torch.inference_mode()
def translate_batch(texts: List[str], src: str, tgt: str, max_len=512):
    tok, mdl = get_translator(src, tgt)
    batch = tok(texts, return_tensors="pt", padding=True, truncation=True, max_length=max_len).to(DEVICE)
    gen = mdl.generate(**batch, max_length=max_len, num_beams=4)
    return tok.batch_decode(gen, skip_special_tokens=True)

def back_translate_batch(texts: List[str], pivot: str, lang="en"):
    to_pivot = translate_batch(texts, lang, pivot)
    back = translate_batch(to_pivot, pivot, lang)
    return back

# Similarity filter
SIM_MODEL = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2", device=DEVICE)

def is_too_similar(a: str, b: str, thresh=0.99):
    emb = SIM_MODEL.encode([a, b], normalize_embeddings=True)
    sim = float(util.cos_sim(emb[0], emb[1]))
    return sim >= thresh

def length_ratio_ok(a: str, b: str, lo=0.7, hi=1.3):
    r = len(b.split()) / (len(a.split()) + 1e-9)
    return lo <= r <= hi

# Augment with progress bar
def augment_minority_df(
    df: pd.DataFrame,
    text_col: str,
    label_col: str,
    minority_label,
    target_size: int,
    pivots: List[str],
    lang="en",
    max_per_source=3,
    batch_size=16
):
    minority_df = df[df[label_col] == minority_label]
    if len(minority_df) >= target_size:
        return pd.DataFrame([])  # nothing to do

    needed = target_size - len(minority_df)
    augmented_rows = []

    for pivot in pivots:
        batches = range(0, len(minority_df), batch_size)
        for start in tqdm(batches, desc=f"Pivot {pivot}", leave=True):
            batch_rows = minority_df.iloc[start:start+batch_size]
            back_texts = back_translate_batch(batch_rows[text_col].tolist(), pivot, lang)

            for orig_row, new_text in zip(batch_rows.itertuples(index=False), back_texts):
                if len(augmented_rows) >= needed:
                    break
                orig_text = getattr(orig_row, text_col)
                if new_text.strip() == orig_text.strip():
                    continue
                if not length_ratio_ok(orig_text, new_text):
                    continue
                if is_too_similar(orig_text, new_text):
                    continue
                new_row = pd.Series(orig_row._asdict())
                new_row[text_col] = new_text
                augmented_rows.append(new_row)

        if len(augmented_rows) >= needed:
            break

    return pd.DataFrame(augmented_rows)
```

```python
# Example usage
aug_df = augment_minority_df(
    df=data_train,
    text_col='cleaned_text',
    label_col='fraudulent',
    minority_label=1,
    target_size=5000,
    pivots=['de', 'fr', 'es', 'id', 'zh', 'ru', 'it'],
    lang='en',
    batch_size=16
)

print(aug_df.shape)
```

```python
df_augmented = pd.concat([data_train, aug_df], ignore_index=True)
```
